```python
df['trouble_asleep'] = df['H4SP5']
df['trouble_asleep'].value_counts()
```

```python
df['trouble_sleep']=df['H4SP5']
df['trouble_sleep']=df['trouble_sleep'].loc[df['trouble_sleep'] !=' ']
df['trouble_sleep']=df['trouble_sleep'].loc[df['trouble_sleep'] !='5']
df['trouble_sleep']=df['trouble_sleep'].loc[df['trouble_sleep'] !='6']
df['trouble_sleep']=df['trouble_sleep'].loc[df['trouble_sleep'] !='8']
```

```python
df['trouble_sleep'].replace(['0','1','2','3','4'],['never in the past four weeks','less than once a week','1 or 2 times a week','3 or 4 times a week','5 or more times a week'],inplace=True)
```
